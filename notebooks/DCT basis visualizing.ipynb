```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
def make_dct_basis1d(N):
    xx = np.arange(0, N)[None, :]
    uu = xx.T
    C = np.cos(np.pi * (2 * xx + 1) * uu / 2 / N)
    return C  # rows are dct basis vectors
```

```python
def make_dct_basis2d(N):
    C = make_dct_basis1d(N)
    basis_imgs = np.zeros((N, N, N, N), dtype=float)
    for i in range(N):
        for j in range(N):
            basis_imgs[i, j] = C[i, :, None] @ C[j, :, None].T
    return basis_imgs
```

```python
def plot_basis(N, cmap="gray"):
    C2 = make_dct_basis2d(N)
    vm = np.max(np.abs(C2))
    fig, axes = plt.subplots(N, N)
    for i in range(N):
        for j in range(N):
            axes[i, j].imshow(C2[i, j], vmin=-vm, vmax=vm, cmap=cmap)
```

```python
plot_basis(8)
```
